## 3.1. List

List là một kiểu dữ liệu dạng tập hợp, nó là một danh sách các phần tử. Nó khá giống với chuỗi kí tự => chuỗi phẩn tử

Phẩn tử ở đây có thể là số, chuỗi, boolean, list,...

```python
thislist = ["apple",123, 123, True, [1,2,3], tuple(), dict(), set()]
print(thislist)
```

### Access List Items

```python
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(thislist[1]) # đếm từ trái qua phải, từ 0 => dương
print(thislist[-1]) # đếm từ phải qua trái, từ 0 => âm 
print(thislist[2:5]) # lấy một đoạn phần từ => sẽ được để trong 1 cái list
a = "apple" in thislist
print(a)
if "apple" in thislist:
  print("Yes, 'apple' is in the fruits list")
```

### Change List Items

```python
thislist = ["apple", "banana", "cherry"]
thislist[1] = "watermelon" 
print(thislist)
```

### List methods

```python
thislist = ["apple", "banana", "cherry"]
thislist.append("orange") # thêm phần tử vào cuối danh sách
print(thislist)
thislist.insert(1, "orange") # thêm phần tử vào chỉ mục chỉ định
print(thislist)
tropical = ["mango", "pineapple", "papaya"]
thislist.extend(tropical) # nối list khác vào sau list hiện tại
print(thislist)
thistuple = ("kiwi", "orange")
thislist.extend(thistuple) # nó còn nối được string, tuple, dict, set,...
print(thislist)
thislist.remove("banana") # xóa phần tử chỉ định
print(thislist)
thislist.pop(1) # xóa phần tử ở chỉ mục chi định
print(thislist)
thislist.clear() # làm trống list
print(thislist)
```

```python
def myfunc(n):
  return abs(n - 50)
            #50, 0, 15, 32, 27
thislist = [100, 50, 65, 82, 23]

thislist.sort(key = myfunc)

print(thislist)
```

```python
thislist = ["banana", "Orange", "Kiwi", "cherry"]
thislist.sort(key = str.lower)
print(thislist)
thislist.reverse()
print(thislist)
```

```python
thislist = ["apple", "banana", "cherry"]
mylist = thislist.copy()
print(mylist)
mylist = list(thislist)
print(mylist)
```

### Loop list

## 3.2. Tuple

Nó là bộ giá trị, có chỉ mục, có thể trùng lặp, nhưng không thể thay đổi giá trị

```python
thistuple = ("apple", "banana", "cherry")
thistuple[1] = "orange"
```

Để thay đổi, có một cách là chuyển nó về list rồi đổi sau đấy lại chuyển về tuple

Và cũng không thể sử dụng append()

```python
thistuple = ("apple", "banana", "cherry")
y = ("orange","noi")
thistuple += y

print(thistuple)
```

Số lượng biến phải khớp với số lượng giá trị trong bộ, nếu không, bạn phải sử dụng dấu hoa thị để thu thập các giá trị còn lại dưới dạng danh sách.

```python
a = ("apple", "mango", "papaya")
(x, y, z) = a
print(x)
print(y)
print(z)
```

```python
fruits = ("apple", "mango")

green, red = fruits

print(green)
print(red)
```

### Dictionaries

Nó là danh sách với những cặp key:value. Nó được sắp xếp theo thứ tự ( từ phiên bản 3.7), có thể thay đổi và không cho phép trung lặp

```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

person = {
  "name": 'Hiệu',
  "age" : 20
}

x = thisdict["model"]
print(x)
x = thisdict.get("model")
print(x)
```

```python
car = {
"brand": "Ford",
"model": "Mustang",
"year": 1964
}

x = car.keys()

print(x) #before the change

car["color"] = "white"

print(x) #after the change

car["year"] = 2000
print(car)

car.update({"year": 2020})
print(car)
```

```python
import pandas as pd
df = pd.read_csv('data/data.csv')
# Cách 1: Dùng vòng lặp để tính freq sau đó dùng sorted để sắp xếp
items = df['itemDescription']
freq_items = {}
'''
{whole milk: 6,
coffee: 1}
'''
for item in items:
    if item in freq_items:
        freq_items[item] += 1
    else:
        freq_items[item] = 1

sorted_freq_items = sorted(freq_items.items(), key = lambda item: item[1], reverse=True)
df_sorted = pd.DataFrame(sorted_freq_items, columns=['items', 'freq'])
print(df_sorted.head(10))
```
