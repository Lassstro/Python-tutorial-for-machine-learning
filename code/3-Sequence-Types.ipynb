### List methods

```python
thislist = ["apple", "banana", "cherry"]
thislist.append("orange") # thm phn t vo cui danh sch
print(thislist)
thislist.insert(1, "orange") # thm phn t vo ch mc ch nh
print(thislist)
tropical = ["mango", "pineapple", "papaya"]
thislist.extend(tropical) # ni list khc vo sau list hin ti
print(thislist)
thistuple = ("kiwi", "orange")
thislist.extend(thistuple) # n cn ni c string, tuple, dict, set,...
print(thislist)
thislist.remove("banana") # xa phn t ch nh
print(thislist)
thislist.pop(1) # xa phn t  ch mc chi nh
print(thislist)
thislist.clear() # lm trng list
print(thislist)
```

```python
import pandas as pd
df = pd.read_csv('data/data.csv')
# Cch 1: Dng vng lp  tnh freq sau  dng sorted  sp xp
items = df['itemDescription']
freq_items = {}
'''
{whole milk: 6,
coffee: 1}
'''
for item in items:
    if item in freq_items:
        freq_items[item] += 1
    else:
        freq_items[item] = 1

sorted_freq_items = sorted(freq_items.items(), key = lambda item: item[1], reverse=True)
df_sorted = pd.DataFrame(sorted_freq_items, columns=['items', 'freq'])
print(df_sorted.head(10))
```
