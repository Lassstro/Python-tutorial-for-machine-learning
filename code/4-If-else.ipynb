Giới thiệu về flowchart trong lập trình

Flowchart trong lập trình Python được sử dụng để biểu diễn các thuật toán hoặc quy trình tính toán trong chương trình Python. Các flowchart được sử dụng để mô tả luồng thực hiện các lệnh, các quyết định và các hành động khác trong chương trình. Flowchart thường được sử dụng để giúp người lập trình có thể hiểu các quy trình tính toán và tìm ra các lỗi trong chương trình.

Các ký hiệu cơ bản được sử dụng trong flowchart bao gồm:

- Hình tròn: Biểu thị cho điểm bắt đầu hoặc kết thúc của một quy trình
- Hình vuông: Biểu thị cho một bước hoặc một lệnh trong quy trình
- Hình thoi: Biểu thị cho một quyết định hoặc một lựa chọn trong quy trình
- Mũi tên: Biểu thị cho hướng đi của quy trình hoặc luồng thực hiện các lệnh

Có nhiều công cụ để vẽ cái này, ví dụ như lucidchart

## Cấu trúc điều kiện lồng nhau

Trong khi thực hiện chương trình, có thể sẽ có những trường hợp chúng ta cần sử dụng cấu trúc if else nhiều lần, các điều kiện lại phụ thuộc nhau. Vì vậy cần sử dụng cấu trúc if else lồng nhau để chương trình gon gàng dễ hiểu và đúng logic

```python
age = 70
if (age > 18):
    print('Trung niên, trung niên, Người già')
    if age > 60:
        print('Người già')
    else:
        print('Trung niên, trung niên')
        if age > 30:
            print('Trung niên')
        else:
            print('Thanh niên')
else:
    print('Trẻ con')
```

Khi cần sử dụng nhiều điều kiện liên tiếp có thể sử dụng elif

```python
age = 10
if age < 6:
    print('Trẻ con')
if  6 <= age < 12 :
    print('Thiếu nhi')
if 12 <= age < 18:
    print('Thiếu niên')
if 18 <= age < 30:
    print('Thanh niên')
if age >= 30 and age < 60:
    print('Trung niên')
if age >= 60:
    print('Người già')
```

```python
age = 20
if age < 6:
    print('Trẻ con')
else:
    if age < 12:
        print('Thiếu nhi')
    else:
        if age < 18:
            print('Thiếu niên')
        else:
            if age < 30:
                print('Thanh niên')
            else:
                if age < 60:
                    print('Trung niên')
                else:
                    print('Người già')
```

Như có thể thấy nó rất rối rắm, vì vậy hãy sử dụng hợp lí hơn bằng cách dùng elif

## elif

Nó tương đương với cấu trúc if else lồng nhau ở trên. Cách nó chạy được giải thích như sau:

```python
age = 10
if age < 6:
    print('Trẻ con')
elif age < 12:
    print('Thiếu nhi')
elif age < 18:
    print('Thiếu niên')
elif age < 30:
    print('Thanh niên')
elif age < 60:
    print('Trung niên')
else:
    print('Người già')
```
