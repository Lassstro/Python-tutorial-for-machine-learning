## Scripts, Modules, Packages in Python?

Đầu tiên là Scripts sẽ là đoạn mã chương trình của chúng ta, khi nó thực thi sẽ thực hiện mong muốn của người lập trình

Khi đoạn mã chương trình của chúng ta có nhiều phần mà quá là dài đi, ta cần phải tách nó ra thành các phần thực hiện các chức năng khác nhau.

Từ đó là có khai niệm Modules. Đây là có đoạn mã được viết riêng ở file khác và khi cần dùng, ta chỉ cần import nó và dùng

Tương tự như thế, package là một gói gồm nhiều module, trong đó thực hiện các chức năng cụ thể

Trong một package có thể gồm nhiều package nhỏ khác

Sau đây là một ví dụ về 1 package tự xây dựng

Khái niệm thư viện trong python khá gần với package. Tuy nhiên có nhiều người cho rằng thư viện là gồm nhiều package

    2 là những thư viện được phát triển bởi bên thứ 3
